```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings(action ='ignore')
```

```python
tips = sns.load_dataset('tips')
```

```python
pd.date_range('2000/1/1', freq = "MS", periods= 5)
```

```python
tips['total_bill'].max()
```

```python
plt.figure(figsize =(7,7))
sns.rugplot(tips['total_bill'],alpha = 0.7)

plt.show()
```

```python
sns.kdeplot(tips['total_bill']);
```

```python
sns.distplot(tips['total_bill'], color ='r',
             rug = True,) # rug 만 False 가 default
             
#              hist = False,
#              kde = False)
plt.show()
```

```python
# seaborn 을 사용한 total_bill 의 boxplot을 하나의 figure에 그리는 수직,수평
plt.figure(figsize = (7,7))

plt.subplot(1,2,1)
sns.boxplot(y = 'total_bill', data= tips)
plt.title('수직')
plt.show() 

plt.subplot(2,2,2)
sns.boxplot(x = 'total_bill', data = tips)
plt.show()
```

```python
# seaborn을 상용 smoker 별  total_bill의 분포를  boxplot을 그린다.
plt.figure(figsize =(7,7))

sns.boxplot(x = 'smoker', y = 'total_bill', data = tips, palette ='rainbow')
plt.show()

```

```python
# seabor을 smoker 별  total_bill의 분포를 성별로  boxplot을 그린다.
plt.figure(figsize =(7,7))

sns.boxplot(x = 'smoker', y = 'total_bill', hue ='sex', data = tips, palette ='rainbow');
```

```python
# 요일별(dat) tip의 분포를 violinplot 을 사용해서 그리기.
# smoker 별로 나누기
# 가로로 그리기 

plt.figure(figsize=(7,7))

sns.violinplot(y = 'tip', x ='day', hue ='smoker',data = tips)


plt.show()

```

```python
plt.figure(figsize=(7,7))

#sns.boxplot(y = tips['tip'], x =tips['smoker'],data = tips)
sns.violinplot(y = 'tip', x ='smoker',data = tips)
sns.swarmplot(y = 'tip' ,x ='smoker', data = tips, palette='rainbow' )

plt.show()
```

```python
# 요일별 흡연자,비흡연자 손님수 (countplot ==bar, barh)
plt.figure(figsize=(7,7))

sns.countplot(x= 'day', hue ='smoker', data = tips )  # 그냥 count가 손님수니까  y 는 설정 안해줘도됨.
```

```python
# 성별 흡연자 
sns.countplot(x= tips['smoker'], hue ='sex', data = tips )
```

```python
sns.countplot(x= 'day', hue ='smoker', data = tips )
```

```python
# scatter plot , linear model(lmplot), #jointplot
```

```python
# scatter plot

# 성별로 나눠서 확인 

plt.figure(figsize =(6,6))

sns.scatterplot(x ='total_bill', y = 'tip', hue = 'sex', data = tips)

plt.show()
```

```python
#linear model(lmplot)

sns.lmplot(x= 'total_bill', y = 'tip', hue='sex' ,data = tips, height = 7)

plt.show()
```

```python
#jointplot

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, palette ='rainbow')
plt.show()
```

```python
# pairplot
sns.pairplot(data = tips, palette ='rainbow')
plt.show()
```
