# Groupby와 집계
## forestfire.csv 파일
- 포르투칼 북서부 산림 산불관련 데이터 셋 [UCI Machine learning repository](https://archive.ics.uci.edu/ml/datasets/forest+fires)    


1. forestfire.csv 파일을 DataFrame으로 읽기
1. 어떤 컬럼들이 범주형 컬럼인지 확인.
1. 월(month)과 요일(day) 의 고유한(unique) 값들과 각각의 행수 조회
1. 요일(day) 별 wind의 평균을 조회
1. 요일(day) 별 wind, rain의 평균을 조회
1. 요일(day) 별 wind, rain의 평균과 합계과 개수 조회
1. 요일(day) 별로 wind는 평균을 rain은 합계와 최대값을 조회
1. 월(month) 그리고 요일별(day) temp의 최대값과 최소값을 조회
1. X좌표(X), Y좌표(Y)별 temp의 평균울 pivot table로 조회


```python
# forestfire.csv 파일을 DataFrame으로 읽기
forestfire_df=pd.read_csv('data2/forestfire.csv')
forestfire_df['X'].unique()#.value_counts
```

```python
# 어떤 컬럼들이 범주형 컬럼인지 확인.
forestfire_df.info()  # 범주형 : month, day,X,Y
```

```python
# 월(month)과 요일(day) 의 고유한(unique) 값들과 각각의 행수 조회
forestfire_df['month'].unique()
```

```python
forestfire_df['month'].value_counts()
```

```python
forestfire_df['day'].value_counts()
```

```python
forestfire_df[['month','day']].value_counts().to_frame()
```

```python
# 요일(day) 별 wind의 평균을 조회
forestfire_df.groupby('day')['wind'].mean()
```

```python
# 요일(day) 별 wind, rain의 평균을 조회
forestfire_df.groupby('day')[['wind','rain']].mean()
```

```python
# 요일(day) 별 wind, rain의 평균과 합계과 개수 조회
forestfire_df.groupby('day')[['wind','rain']].agg(['mean','sum','count'])
```

```python
# 요일(day) 별로 wind는 평균을 rain은 합계와 최대값을 조회
forestfire_df.groupby('day').agg({'wind':'mean','rain':['sum','max']})
```

```python
# 월(month) 그리고 요일별(day) temp의 최대값과 최소값을 조회
ffdf=forestfire_df.groupby(['month','day'])['temp'].agg(['max','min'])  #--> 저장해서 파일로 보면 전체 데이터 한번에 다볼 수 있다.
ffdf
```

```python
for row in ffdf.iterrows():  # DataFrame을 한행씩 제공하는 generator
    print(row)
```

```python
forestfire_df.pivot_table(values ='temp', index = 'month', columns ='day', aggfunc=['max','min'])
```

```python
# X좌표(X), Y좌표(Y)별 temp의 평균울 pivot table로 조회
```

```python
forestfire_df.groupby(['X','Y'])['temp'].mean()
```

```python
forestfire_df.pivot_table(values= 'temp', index ='X', columns='Y', aggfunc ='mean')
```
