```python
# s_2016,s_2017 --> join innerjoin
pd.concat([s_2016,s_2017], axis = 1,join = 'inner')
```

```python
#s_2017 과 s_info 인덱스를  Symbol 로 바꾸고 concat axis=1
pd.concat([s_2017.set_index('Symbol'),s_info.set_index('Symbol')], axis =1, join = 'inner')
```

```python
# 2016,2017  join사용하여 join
# 순서 바꿔서도 해보기
s_2016.join(s_2017,lsuffix ='_2016')
```

```python
# 컬럼명에  suffix를 붙인다.  3개이상 join 하는데 컬럼 명이 중복이 계속 될때 !!! 2개는 lsuffix rsuffix 쓰ㅡ면된다.
s_2017.add_suffix('_2017').join([s_2016.add_suffix('_2016'),
                                s_2018.add_suffix('_2018'),
                                s_info.add_suffix('_info')])
                                    
```

```python
#join(others): others에 DataFrame 들의 list를 전달하는 경우에는 (여러개 조인) right join은 안된다.
```

```python
#s_2018 과 s_info 인덱스를  Symbol 로 바꾸고 join 이용
s_2018.set_index('Symbol').join(s_info.set_index('Symbol'))
```

```python
#2016,2018 join suffixes 도 정하기
s_2016.merge(s_2018, on= 'Symbol')
```

```python
s_2016.merge(s_2018, on= 'Symbol', suffixes=['_2016','_2018'])
```

```python
# 같은 컬럼이 없을때!(2018_copy의 Symbol을 index명으로 뺏기때문에 같은컬럼이 없어짐)  // 2018_copy의 인덱스명과 info컬럼명이 같을때 
s_2018_copy = s_2018.append({'Symbol':'AAA','Shares':5,'Low':100,'High':120},ignore_index = True)
s_2018_copy
```

```python
s_2018_copy.set_index('Symbol',inplace =True)
```

```python
# 같은 컬럼이 없을때!(2018_copy의 Symbol을 index명으로 뺏기때문에 같은컬럼이 없어짐)  // 2018_copy의 인덱스명과 info컬럼명이 같을때 
s_info.merge(s_2018_copy, left_on = 'Symbol', right_index =True, how = 'outer' )
```
