```python
# max  min      
def max_min_diff(x):  #Series  DataFrame  ()
    if x.dtype == 'object':
        return None
    return x.max()-x.min()
```

```python
flights['ARR_DELAY'].head()  #  .
```

```python
flights['ARR_DELAY'].agg(max_min_diff)
```

```python
max_min_diff(flights['ARR_DELAY'])
```

```python
flights['ARR_DELAY'].agg(['min','max',max_min_diff])
```

```python
flights.agg(['max','min',max_min_diff]) #   
```

```python
flights.groupby('AIRLINE').agg(['min',max_min_diff])
```

```python
def check_max(series,n=1000):
    if series.dtype != 'object':
        return series.max() >= n  # n  boolean  True, False  
```

```python
flights['ARR_DELAY'].agg(check_max,n=1000)
```

```python
flights['ARR_DELAY'].agg(check_max, n= 2000)
```

```python
flights.groupby('AIRLINE')['ARR_DELAY'].agg(check_max,n=302)
```
